```python
import re
def preprocess(text):
  tt=re.sub(r'[^A-Za-z0-9]+'," ",text)
  tt= re.sub(r'#\w+',"",text)
  tt= re.sub(r'@\w+',"",text)#links ,special character
  tt=re.sub(r'<.*?>'," ",text)#html 
  tt=re.sub(r'https\S+'," ",text)#website 
  return tt

preprocess("@hbhbhbhbu </html> 190000: {{ }: {>>< >< jn jn fu n un f https://colab.research.google.com/drive/1E388UgfXma-vh2MdAY3hmYCmcbvj0BS5#scrollTo=ohru7TnPSGh8")
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()

def text_preprocess(text):
  l1=[]
  #tt=text.lower()
  tt=nltk.word_tokenize(text.lower())
  tt=[lemmatizer.lemmatize(word) for word in tt if word not in set(stopwords.words('english'))]
  l1.append(tt)
  return (l1)
```

```python
df['cclean'] =df['Cleaned_text'].apply(lambda x: text_preprocess(x))
df.head()
```

```python
import gensim
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (df['clean'], df['user_suggestion'] , test_size=0.2)
```

```python
model = gensim.models.Word2Vec(X_train,size=100,window=5,min_count=2)
```

```python
model.wv.most_similar('jump')
```

```python
model.wv.vocab
```

```python
words=(model.wv.vocab)
X_train_vect = np.array([np.array([model.wv[i] for i in ls if i in words]) for ls in X_train])

X_test_vect = np.array([np.array([model.wv[i] for i in ls if i in words])
                         for ls in X_test])
```

```python
X_train_vect_avg = []
for v in X_train_vect:
    if v.size:
        X_train_vect_avg.append(v.mean(axis=0))
    else:
        X_train_vect_avg.append(np.zeros(100, dtype=float))
        
X_test_vect_avg = []
for v in X_test_vect:
    if v.size:
        X_test_vect_avg.append(v.mean(axis=0))
    else:
        X_test_vect_avg.append(np.zeros(100, dtype=float))
```

```python
model = Sequential()
model.add(Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train_vect_avg, y_train.values.ravel(), epochs=3, batch_size=64)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenise=Tokenizer(1522)
tokenise.fit_on_texts(df['clean'])
```

```python
seq=tokenise.texts_to_sequences(df['clean'].values)
len(seq)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded=pad_sequences(seq,100)
```

```python
num_words = 1522
embedding_matrix = np.zeros((num_words, 100))
for word, i in tokenise.word_index.items():
    if i >= num_words:
        continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```
